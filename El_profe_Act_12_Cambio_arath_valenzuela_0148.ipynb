## Las variables
1. CP --> Costo del producto
2. CD --> Cantidad de dinero
3. Cambio

## Formulas
$Cambio=CD-CP$

## Diagrama de flujo en LucidChart
<img src="https://raw.githubusercontent.com/JacoboValenzuela128/Calcular-cambio-if...else-/refs/heads/main/DF%20if%20else%20cambio.png" width="350" height="500" alt="Texto descriptivo" />

```python
# Para cuando es verdadero (si)
# Calculando el cambio usando if else
# Datos de entrada: CD, CP
# Mostrar Cambio si es falso
# Si es verdadero mostrar ´no alcanza´
CP = int(input("Ingrese el costo del producto: "))
CD = int(input("Ingrese la cantidad de dinero: "))
if CP > CD:
    print("No alcanza")
else:
    Cambio = CD - CP
    print("El cambio es: ", Cambio)

print("Fin del programa")
print("Programa realizado por Arath Valenzuela NL:52")
```

```python
# Para cuando es falso (no)
# Calculando el cambio usando if else
# Datos de entrada: CD, CP
# Mostrar Cambio si es falso
# Si es verdadero mostrar ´no alcanza´
CP = int(input("Ingrese el costo del producto: "))
CD = int(input("Ingrese la cantidad de dinero: "))
if CP > CD:
    print("No alcanza")
else:
    Cambio = CD - CP
    print("El cambio es: ", Cambio)

print("Fin del programa")
print("Programa realizado por Arath Valenzuela NL:52")
```
